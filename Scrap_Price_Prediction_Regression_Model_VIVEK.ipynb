```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
X['name']=le.fit_transform(X['name'])
X['fueltypes']=le.fit_transform(X['fueltypes'])
X['aspiration']=le.fit_transform(X['aspiration'])
X['doornumbers']=le.fit_transform(X['doornumbers'])
X['carbody']=le.fit_transform(X['carbody'])
X['drivewheels']=le.fit_transform(X['drivewheels'])
X['enginelocation']=le.fit_transform(X['enginelocation'])
X['enginetype']=le.fit_transform(X['enginetype'])
X['cylindernumber']=le.fit_transform(X['cylindernumber'])
X['fuelsystem']=le.fit_transform(X['fuelsystem'])
X
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X_std=std.fit_transform(X)
X_std
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split (X_std, Y, test_size = 0.2, random_state = 23)  # 80:20 , 70:30, # parado principle
x_train, x_test, y_train, y_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
std_reg = LinearRegression()
std_reg.fit(x_train, y_train)
```

```python
y_predict = std_reg.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
MAE=mean_absolute_error(y_predict,y_test)
MSE=mean_squared_error(y_predict,y_test)
RMSE=np.sqrt(MSE)
R_Squared=r2_score(y_predict,y_test)
```

```python
print("MAE is" ,MAE,'\n' "MSE is" ,MSE,'\n' "RMSE is" ,RMSE,'\n' "R_Squared value is" ,R_Squared)
```

```python
plt.scatter(y_test,y_predict)
plt.plot((y_test,y_predict),color='red',linestyle='-')
plt.title("Car price prediction")
plt.xlabel("Actual price")
plt.ylabel("predicted price")
plt.show()
```
